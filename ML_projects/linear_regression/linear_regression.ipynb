```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv('d:\Projects\data\small_df.csv')
df=pd.DataFrame(data)
df["index_col"] = range(1, len(df) + 1)


 
x=df["Original Price"]
y=df['Current Price']
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)



Y=np.array(y_train).reshape([-1,1])
X=np.column_stack((np.ones((len(X_train))),X_train))
THETA=np.ones((2,1))

def loss_fxn(X,Y,THETA,lam):
    return np.mean((X @ THETA- Y)**2) + lam * np.vstack(([0],THETA[1:]))

def solutions(X,Y,lam):
    return np.linalg.inv(X.T @ X + len(df) * lam * np.eye(2)) @ X.T @ Y

lam=1200
THETA=solutions(X,Y,lam)
x=np.linspace(np.min(X),np.max(X),10)
y= THETA[1] * x + THETA[0] 
print(loss_fxn(X,Y,THETA,lam))
plt.scatter(X[:,1],Y) 
plt.plot(x,y)    
plt.show()
print(THETA)
```

```python
y_pred=THETA[1] * X_test + THETA[0]
r2_score(y_test,y_pred)
```
